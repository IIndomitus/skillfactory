```python
# Функция для перебора автомобильных марок.
def mark_iter(area):
    for mark in all_marks_list:
        print(mark)
        mark_search(mark, area)
```

```python
# Функция для непосредственного поиска информации в запрошенных данных и сохранения найденной информации.
def get_info(data):
    for i in range (1, len(data)):
        try:
            bodyType = data[i]['vehicle_info']['configuration']['body_type']
        except:
            bodyType = None 

        try:
            brand = data[i]['vehicle_info']['mark_info']['code']
        except:
            brand = None
    
    
        try:
            color = data[i]['color_hex']
        except:
            color = None


        try:
            description = data[i]['description']
        except:
            description = None

        try:
            engineDisplacement = data[i]['vehicle_info']['tech_param']['displacement']
        except:
            engineDisplacement = None

        try:
            enginePower = data[i]['vehicle_info']['tech_param']['power']
        except:
            enginePower = None

        try:
            equipment_dict = data[i]['vehicle_info']['equipment']
        except:
            equipment_dict = None

        try:
            fuelType = data[i]['vehicle_info']['tech_param']['engine_type']
        except:
            fuelType = None

        try:
            mileage = data[i]['state']['mileage']
        except:
            mileage = None

        try:
            modelDate = data[i]['vehicle_info']['super_gen']['year_from']
        except:
            modelDate = None

        try:
            model_info = data[i]['vehicle_info']['model_info']
        except:
            model_info = None

        try:
            model_name = data[i]['vehicle_info']['model_info']['name']
        except:
            model_name = None

        try:
            name = (data[i]['vehicle_info']['mark_info']['code'] + " " 
                    + data[i]['vehicle_info']['model_info']['name'] +  " " 
                    + data[i]['vehicle_info']['super_gen']['name'])
        except:
            name = None

        try:
            numberOfDoors = data[i]['vehicle_info']['configuration']['doors_count']
        except:
            numberOfDoors = None

        try:
            price = data[i]['price_info']['price']
        except:
            price = None


        try:
            productionDate = data[i]['documents']['year']
        except:
            productionDate = None

        try:
            vehicleConfiguration = data[i]['vehicle_info']['configuration']['body_type']
        except:
            vehicleConfiguration = None

        try:
            vehicleTransmission = data[i]['vehicle_info']['tech_param']['transmission']
        except:
            vehicleTransmission = None

        try:
            vendor = data[i]['vehicle_info']['vendor']
        except:
            vendor = None

        try:
            owners_number = data[i]['documents']['owners_number']
        except:
            owners_number = None

        try:
            ownership = data[i]['documents']['year']
        except:
            ownership = None

        try:
            pts = data[i]['documents']['pts']
        except:
            pts = None

        try:
            gear_type = data[i]['vehicle_info']['tech_param']['gear_type']
        except:
            gear_type = None

        try:
            steering_wheel = data[i]['vehicle_info']['steering_wheel']
        except:
            steering_wheel = None 

        try:
            repair = None
        except:
            repair = None

        try:
            custom = data[i]['documents']['custom_cleared']
        except:
            custom = None


        with open(f"new_train.csv", "a", encoding='utf-8') as file:
            writer = csv.writer(file)

            writer.writerow(
                (str(bodyType),
                 str(brand),
                 str(color),
                 str(description),
                 str(engineDisplacement),
                 str(enginePower),
                 str(equipment_dict),
                 str(fuelType),
                 str(mileage),
                 str(modelDate),
                 str(model_info),
                 str(model_name),
                 str(name),
                 str(numberOfDoors),
                 str(price),
                 str(productionDate),
                 str(vehicleConfiguration),
                 str(vehicleTransmission),
                 str(vendor),
                 str(owners_number),
                 str(ownership),
                 str(pts),
                 str(gear_type),
                 str(steering_wheel),
                 str(repair),
                 str(custom))
            )
```
